```python
Male_df = df[df['gender'] == 'male']
Male_df
```

```python
Female_df = df[df['gender'] == 'female']
Female_df
```

```python
df_Male_none = Male_df[Male_df['test preparation course'] == 'none']
df_Male_comp = Male_df[Male_df['test preparation course'] == 'completed']
data = [['prepared',df_Male_comp['gender'].count()],['not prepared',df_Male_none['gender'].count()]]
df_q5 = pd.DataFrame(data,columns = ['Test Prep','Number'])
df_q5
```
